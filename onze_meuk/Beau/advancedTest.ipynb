```python
import rebound
sim = rebound.Simulation()
sim.add(m=1.) #Star A
sim.add(m=0.9, a=2.5) #Star B
sim.add(m=0.001, a=0.8, primary=sim.particles[0]) #Planet Aa
sim.add(m=0.001, a=0.8, primary=sim.particles[1]) #Bb, 
sim.move_to_com()
fig = rebound.OrbitPlot(sim)
```

```python
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
sim.move_to_com()
for i in range(300):
    sim.integrate(sim.t+0.22)
    fig, ax = rebound.OrbitPlot(sim,color=True,unitlabel="[AU]",xlim=[-3,3.],ylim=[-4,3.])
    display(fig)
    plt.close(fig)
    clear_output(wait=True)
```
