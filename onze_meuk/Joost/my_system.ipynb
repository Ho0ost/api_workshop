```python
import rebound
import matplotlib.pyplot as plt
import numpy as np
```

```python
sim = rebound.Simulation()
sim.units = ('AU', 'yr', 'Msun')
sim.add("Sun")
#sim.add("Jupiter")

sim.integrator = "whfast"
sim.dt = .01
```

```python
for j in range (100):
    for i in range (100):
        mass=1e-9
        radius=1+0.001*j
        omega=(2*np.pi)/100*i
        sim.add(m=mass,a=radius,omega=omega)
```

```python
fig = rebound.OrbitPlot(sim)
```
