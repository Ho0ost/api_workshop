```python
import numpy as np
import time
import rebound
import os
sim = rebound.Simulation()

initial_distance = 0.3
sim.add(m=1)
for i in range (5):
    sim.add(m=1e-3, a=initial_distance+(i*initial_distance))


sim.move_to_com() # Move to the centre of mass frame


w = sim.getWidget()
w
sim.getWidget(size=(500,500))
```

```python
sim_lenght = 100
print(sim_lenght*2*np.pi,"=",sim_lenght," years")

for i in range (sim_lenght):
    sim.integrate(i*2*np.pi)
    time.sleep(0.1)

    
```

```python
orbits = sim.calculate_orbits()
for orbit in orbits:
    print(orbit)
```
